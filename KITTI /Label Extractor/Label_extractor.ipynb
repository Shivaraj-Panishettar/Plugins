```python
import os

def extract_bounding_box(data):
    # Split the data by spaces
    parts = data.split()
    
    # Check if there are at least 8 elements (0-based index 7 is the end of bounding box coordinates)
    if len(parts) >= 8:
        # Extract the bounding box values from indices 4, 5, 6, and 7
        bbox = f"{parts[4]} {parts[5]} {parts[6]} {parts[7]}"
        return bbox
    return None

def save_bounding_box(output_dir, filename, bbox):
    output_path = os.path.join(output_dir, f"{os.path.splitext(filename)[0]}_bbox.txt")
    with open(output_path, 'w') as f:
        f.write(bbox + '\n')

def process_files(input_dir, output_dir):
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)
    
    for filename in os.listdir(input_dir):
        if filename.endswith('.txt'):  # Assuming all your files are text files
            file_path = os.path.join(input_dir, filename)
            with open(file_path, 'r') as file:
                data = file.read().strip()
                
                # Extract bounding box
                bbox = extract_bounding_box(data)
                
                if bbox:
                    # Save the bounding box data to a new file
                    save_bounding_box(output_dir, filename, bbox)
                    print(f"Bounding box data saved for {filename}")
                else:
                    print(f"No bounding box data found in {filename}")


# Define your input and output directories
input_directory = 'C:/Users/cim-security/Desktop/archive/training/label_2'
output_directory = 'C:/Users/cim-security/Desktop/archive/training/Label'

# Process all files in the input directory
process_files(input_directory, output_directory)
```

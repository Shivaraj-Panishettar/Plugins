```python
import numpy as np
import struct
import os
import open3d as o3d

def bin_to_pcd(binFileName, include_intensity):
    size_float = 4
    list_pcd = []
    with open(binFileName, "rb") as f:
        byte = f.read(size_float * 4)
        while byte:
            x, y, z, intensity = struct.unpack("ffff", byte)
            list_pcd.append([x, y, z])
            if include_intensity:
                # Optionally, you can do something with intensity if needed
                pass
            byte = f.read(size_float * 4)
    
    np_pcd = np.asarray(list_pcd)
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(np_pcd)
    return pcd

def main(input_folder, output_folder, include_intensity):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for file_name in os.listdir(input_folder):
        if file_name.endswith('.bin'):
            bin_file_path = os.path.join(input_folder, file_name)
            pcd_file_name = file_name.replace('.bin', '.pcd')
            pcd_file_path = os.path.join(output_folder, pcd_file_name)
            
            pcd = bin_to_pcd(bin_file_path, include_intensity)
            o3d.io.write_point_cloud(pcd_file_path, pcd)
            print(f"Converted {file_name} to {pcd_file_name}")

# Specify your input and output folders here
input_folder = "S:/KITTI/training/velodyne"  # Update with your BIN folder path
output_folder = "S:/KITTI/training/velodyne_pcd"  # Update with your desired output folder path
include_intensity = True  # Set to False if you do not want to include intensity

# Run the conversion
main(input_folder, output_folder, include_intensity)
```
